```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.data[:10]
```

```python
iris.target_names[iris.target]
```

```python
iris.feature_names
```

```python
import matplotlib.pyplot as plt

plt.scatter(iris.data[:,1], iris.data[:,2], c=iris.target)
plt.xlabel(iris.feature_names[1])
plt.ylabel(iris.feature_names[2])
plt.show()
```

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier(max_depth=3)
clf = clf.fit(iris.data, iris.target_names[iris.target])

clf.score(iris.data, iris.target_names[iris.target])
```

```python
plt.figure(figsize=(15, 10))
tree.plot_tree(clf, feature_names=iris.feature_names , class_names=iris.target_names)
plt.show()
```

```python
# save model
from joblib import dump, load
dump(clf, 'my_model.joblib') 
```
